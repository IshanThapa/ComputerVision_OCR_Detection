```python
import os
folder_path = "/kaggle/input/aadhar/QA - 5 - Aadhar Cards"
files = []
main_files = os.listdir(folder_path)
for file_name in main_files:
    file_path = os.path.join(folder_path, file_name)
    files.append(file_path)
```

```python
import pytesseract
from PIL import Image
def extract_text_tesseract(image_path):
    return pytesseract.image_to_string(Image.open(image_path), lang='hin+eng')

all_data = []
for image_path in files:
    text_tesseract = extract_text_tesseract(image_path)
    all_data.append(text_tesseract)
```

```python
len(all_data)
```

```python
all_result = []
for data in all_data:
    result = get_result(data)
    all_result.append(result)
```

```python
all_result = [eval(data) for data in all_result]
```

```python
import pandas as pd
df = pd.DataFrame(all_result)
df["file"] = main_files
df.head()
```
